# INICIO DO TRATAMENTO DO DATABASE NO SQLITE

```python
import sqlite3
import csv

# Conectar ao banco de dados SQLite (substitua 'seu_banco_de_dados.db' pelo nome do seu banco de dados)
conn = sqlite3.connect('pf_maternidade2.db')
cursor = conn.cursor()

file = open("pf_maternidade.csv", encoding='UTF-8')
conteudo = csv.reader(file)

# Criar a tabela com nomes de colunas mais curtos
cursor.execute("""
    CREATE TABLE IF NOT EXISTS maternidade (
        "ID" INTEGER PRIMARY KEY,
        "Carimbo" TEXT,
        "UF" TEXT,
        "Idade" INTEGER,
        "NumFilhos" INTEGER,
        "Escolaridade" TEXT,
        "Renda" TEXT,
        "IdadeGravidez" TEXT,
        "AnoFilho" TEXT,
        "EstudandoGravidez" TEXT,
        "InterrupcaoEstudos" TEXT,
        "TempoInterrupcao" TEXT,
        "GravidezProgramada" TEXT,
        "Dificuldade" TEXT,
        "LocalCuidadores" TEXT
    )
""")

# Inserir dados na tabela
cursor.executemany("""
    INSERT INTO maternidade 
    ("Carimbo", "UF", "Idade", "NumFilhos", "Escolaridade", 
     "Renda", "IdadeGravidez", "AnoFilho", 
     "EstudandoGravidez", "InterrupcaoEstudos", "TempoInterrupcao", 
     "GravidezProgramada", "Dificuldade", "LocalCuidadores")
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
""", conteudo)

# Adicionar a coluna de ID sequencial
cursor.execute("""
    CREATE TABLE IF NOT EXISTS maternidade_temp AS 
    SELECT 
        ROW_NUMBER() OVER (ORDER BY "Carimbo") AS ID,
        "Carimbo", "UF", "Idade", "NumFilhos", "Escolaridade", 
        "Renda", "IdadeGravidez", "AnoFilho", 
        "EstudandoGravidez", "InterrupcaoEstudos", "TempoInterrupcao", 
        "GravidezProgramada", "Dificuldade", "LocalCuidadores"
    FROM maternidade
""")
cursor.execute("""
    DROP TABLE maternidade
""")
cursor.execute("""
    ALTER TABLE maternidade_temp
    RENAME TO maternidade
""")

# Confirmar a alteração
conn.commit()

# Fechar a conexão
conn.close()


```

Excluindo coluna 'Carimbo'

```python
def excluir_coluna(nome_coluna):
    # Conectar ao banco de dados SQLite
    conn = sqlite3.connect('pf_maternidade2.db')
    cursor = conn.cursor()

    # Obter as colunas originais da tabela
    cursor.execute(f"PRAGMA table_info(maternidade)")
    colunas_originais = [coluna[1] for coluna in cursor.fetchall()]

    # Verificar se a coluna desejada existe
    if nome_coluna not in colunas_originais:
        print(f"A coluna '{nome_coluna}' não existe na tabela.")
        conn.close()
        return

    # Criar uma nova tabela sem a coluna desejada
    cursor.execute(f"""
        CREATE TABLE IF NOT EXISTS maternidade_temp AS
        SELECT {", ".join(colunas for colunas in colunas_originais if colunas != nome_coluna)} FROM maternidade
    """)

    # Remover a tabela original
    cursor.execute("DROP TABLE IF EXISTS maternidade")

    # Renomear a nova tabela para o nome original
    cursor.execute("ALTER TABLE maternidade_temp RENAME TO maternidade")

    conn.commit()
    conn.close()


excluir_coluna('Carimbo')

```

PADRONIZAÇÃO ESTADOS

```python
import sqlite3
import csv

def padronizar_estados(conn):
    cursor = conn.cursor()
    
    # Mapeamento de estados
    mapeamento_estados = {
        'sao paulo': 'São Paulo',
        'santa catarina': 'Santa Catarina',
        'bahia': 'Bahia',
        'pernambuco': 'Pernambuco',
        'distrito federal': 'Brasília',
        'rio de janeiro': 'Rio de Janeiro',
        'ceara': 'Ceará',
        'paraiba': 'Paraíba',
        'piaui': 'Piauí',
        'Santa Catarina/ SC': 'Santa Catarina',
        'SP Santo André': 'São Paulo',
        'São Paulo / SP': 'São Paulo',
        'São  Paulo': 'São Paulo',
        'São Paulo': 'São Paulo',
        'São Paulo/ SP' : 'São Paulo', 
        'Valparaíso de Goiás -GO' : 'Goiás',
        'Paraíba/BR' : 'Paraíba',
        'SP': 'São Paulo',
        'Sp': 'São Paulo',
        'Sc': 'Santa Catarina',
        'RJ': 'Rio de Janeiro',
        'DF': 'Brasília',
        'PE': 'Pernambuco'
    }
    

    # Adicionar uma coluna "País" com o valor "Brasil" para todas as linhas
    cursor.execute("ALTER TABLE maternidade ADD COLUMN Pais TEXT DEFAULT 'Brasil'")

    # Atualizar estados na tabela
    for estado_antigo, estado_novo in mapeamento_estados.items():
        cursor.execute(f"UPDATE maternidade SET UF = '{estado_novo}' WHERE UF LIKE '{estado_antigo}'")

    conn.commit()

def salvar_para_csv(conn, nome_arquivo):
    cursor = conn.cursor()

    # Recuperar dados da tabela maternidade
    cursor.execute("SELECT * FROM maternidade")
    dados = cursor.fetchall()

    # Obter os nomes das colunas
    colunas = [desc[0] for desc in cursor.description]

    # Escrever os dados para um arquivo CSV
    with open(nome_arquivo, 'w', newline='', encoding='utf-8') as arquivo_csv:
        escritor_csv = csv.writer(arquivo_csv)
        
        # Escrever cabeçalho
        escritor_csv.writerow(colunas)
        
        # Escrever os dados
        escritor_csv.writerows(dados)

def main():
    conn = sqlite3.connect('pf_maternidade2.db')

    # Aplicar as alterações
    padronizar_estados(conn)

    # Salvar os dados em um arquivo CSV
    salvar_para_csv(conn, 'dados_maternidade.csv')

    conn.close()

if __name__ == "__main__":
    main()

```
