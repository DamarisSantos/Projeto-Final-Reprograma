Todos os anos milhares de mulheres abandonam os estudos por diversos motivos, sendo a falta de apoio das instituições de ensino uma das principais quando elas se tornam mães.
Atualmente existem aguns projetos de lei em andamento para que mude essa realidade, porém são todos muito recentes, e ainda podem percorrer um longo caminho até se tornar lei de fato.
Devido a ausência de dados senti a necessidade de elaborar uma pesquisa, após realizar a análise, sinto falta de algumas perguntas, mas já é possível obter alguns dados interessantes sobre o assunto.


```python
#IMPORTANDO AS BIBLIOTÉCAS

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pf_maternidade.csv", sep="," , encoding= 'UTF-8') 
df
```

```python
df = df.rename(columns={'Carimbo de data/hora': 'Data'})
df
```

```python
df.shape
```

```python
# Contagem de respostas
contagem_respostas = df['VOCÊ INTERROMPEU OU ADIOU OS ESTUDOS POR CONTA DO NASCIMENTO DO FILHO(A)? devem ser considerados também cursos livres / pós graduação'].value_counts()

# Crie um gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(contagem_respostas, labels=contagem_respostas.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Interrupção ou Adiamento de Estudos por Conta do Nascimento do Filho(a)')
plt.show()
```

No gráfico acima verifica-se que aproximadamente 31% das mulheres adiaram ou pausaram os estudos por ocorrência do nascimento do filho.
O que é um número considerável

```python
# Contagem de respostas
contagem_respostas2 = df['POR QUANTO TEMPO VOCÊ INTERROMPEU OU ADIOU OS ESTUDOS?'].value_counts()

plt.figure(figsize=(10, 6))
contagem_respostas2.plot(kind='bar', color='skyblue', edgecolor='black', linewidth=1.2)

# título
plt.xlabel('Tempo de Adiamento dos Estudos')
plt.ylabel('Número de Respostas')
plt.title('Tempo de Adiamento dos Estudos')

# percentual
for i, value in enumerate(contagem_respostas2):
   plt.text(i, value + 0.1, f'{value / len(df) * 100:.1f}%', ha='center')

plt.show()
```

No gráfico acima verifica-se que a maioria das mulheres precisou adiar os estudos por mais de 2 anos ou nunca retornaram.
